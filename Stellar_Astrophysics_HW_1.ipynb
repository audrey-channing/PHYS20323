```python
#Problem 1
import numpy as np
import matplotlib.pyplot as plt
#part a
#constants
radius_venus_cm = 6051.8 * 1e5 # Radius of Venus in cm (1 km = 1e5 cm)
au_to_cm = 1.496e+13  # 1 AU in cm
radius_venus_au = radius_venus_cm / au_to_cm # Radius of Venus in AU

#distance ranges
closest_distance= 0.28  #closest distance Venus will get to earth in AU
furthest_distance= 1.72 #furthest distance Venus will get to earth in AU
d_step = 0.1            # Step size in AU
distances = np.arange(closest_distance, furthest_distance + d_step, d_step) #storing each distance with the desiginated step size

#Equation for calculating the solid angle 
solid_angles_equation = np.pi * (radius_venus_au **2) / (distances**2)

#Plot to show distance range
plt.figure(figsize=(8, 6))
plt.plot(distances, solid_angles_equation, marker='o', linestyle='-', color='blue', label='Solid Angle')
plt.xlabel('Earth-Venus Distance (AU)', fontsize=12)
plt.ylabel('Solid Angle (steradians)', fontsize=12)
plt.title('Solid Angle vs. Earth-Venus Distance', fontsize=14)
plt.grid(True, linestyle='--', alpha=0.7)
plt.legend(fontsize=12)
plt.tight_layout()
plt.show()
#######################################################
#part b

min_solid_angle = np.min(solid_angles_equation)
max_solid_angle = np.max(solid_angles_equation)

#calculating the percent difference between the minimum and maximum solid angle
print('The percent difference between the maximum and minimum solid angle is',((max_solid_angle - min_solid_angle) / max_solid_angle) * 100)

# Total sky solid angle (4 * pi steradians)
total_sky_solid_angle = 4 * np.pi 

# Fraction of the total sky for min and max solid angles
fraction_min = min_solid_angle / total_sky_solid_angle
fraction_max = max_solid_angle / total_sky_solid_angle

print("Fraction of the total sky (min solid angle):", fraction_min)
print("Fraction of the total sky (max solid angle):", fraction_max)
print("The two most extreme solid angles represent the apparent size of Venus in the sky as seen from Earth at its closest and farthest distance to earth.")

########################################################
#part c
luminosity_venus = 5e23  # Visible luminosity of Venus in ergs/s

# Calculate fluxes (F = L / (4 * pi * d^2)) where F is flux, L is luminosity and d is diameter
flux_closest = luminosity_venus / (4 * np.pi * (closest_distance * au_to_cm)**2)
flux_farthest = luminosity_venus / (4 * np.pi * (furthest_distance * au_to_cm)**2)
flux_ratio = flux_closest / flux_farthest

# Calculate intensities (I = F / \Omega) where I is intensity and omega is the solid angle
intensity_closest = flux_closest / max_solid_angle
intensity_farthest = flux_farthest / min_solid_angle
intensity_ratio = intensity_closest / intensity_farthest

print("Flux ratio (closest/farthest):", flux_ratio)
print("Intensity ratio (closest/farthest):", intensity_ratio)
print("The ratios of flux and intensity are different because intensity includes the angular size of Venus, adding an additional scaling factor")
```
